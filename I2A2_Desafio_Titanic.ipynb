## &#9776; **DATA CLEANING**

Antes de iniciarmos essa etapa, vale a pena descrever os tipos de mecanismos de dados faltantes, de forma a compreendermos quais aes a serem tomadas quando da existncia da ausncia de dados em nossos *datasets*. Abaixo, segue os principais mecanismos e suas descries:

a) **MCAR  Missing Completely At Random** : significa que no h relao entre a falta de dados e quaisquer valores, observados ou ausentes. Esses pontos de dados ausentes so um subconjunto aleatrio dos dados. No h nada sistemtico acontecendo que torne alguns dados mais provveis de serem perdidos do que outros.

b) **MAR  Missing At Random (Perdidos Aleatoriamente)** : significa que h uma relao sistemtica entre a propenso de valores ausentes e os dados observados , mas no os dados ausentes. 

c) **MNAR  Missing Not At Random** : significa que h uma relao entre a propenso de um valor estar ausente e seus valores. Este  um caso em que as pessoas com menor escolaridade no tm acesso aos estudos ou as pessoas mais doentes tm maior probabilidade de desistir do estudo.

## &#10074; **PREENCHENDO OS VALORES AUSENTES DE IDADE**

```python
# preenchendo os dados ausentes no atributo age com valores gerados aleatoriamente
# altera o atributo de float64 para int64 
cj_dados = [df_train, df_test]
for cj_dado in cj_dados:
    media = cj_dado["age"].mean()
    desvio_padrao = cj_dado["age"].std()
    is_null = cj_dado["age"].isnull().sum()
    # gerar nmeros aleatrios entre a mdia, std e is_null
    rand_idade = np.random.randint(media - desvio_padrao, media + desvio_padrao, size = is_null)
    # preenche os valores ausentes
    idade_slice = cj_dado["age"].copy()
    idade_slice[np.isnan(idade_slice)] = rand_idade
    cj_dado["age"] = idade_slice
    cj_dado["age"] = cj_dado["age"].astype(int)
```

```python
# verifica se h ainda valores ausentes no atributo age em train e test
print(f"Quantidade de valores ausentes no atributo age  em train: {df_train.age.isnull().sum()}")
print(f"Quantidade de valores ausentes no atributo age  em test: {df_test.age.isnull().sum()}")
```

```python
# cria funo plot para apresentar distribuio
def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , shade= True )
    facet.set( xlim=( 0 , df[ var ].max() ) )
    facet.add_legend()
```

## &#10074; **SELEO DOS ATRIBUTOS DE INTERESSE**

Quanto mais recursos um modelo tiver, mais sensvel ele ser a erros devido  variao. Assim, queremos selecionar os recursos mnimos necessrios para produzir um modelo vlido.

Uma abordagem comum para eliminar recursos  eliminar aqueles que so menos importantes para o modelo.

A seleo de recursos pode ser feita de vrias maneiras, mas existem 3 categorias:

- Mtodo de filtro

- Mtodo envoltrio

- Mtodo incorporado

Nesta etapa, vamos selecionar os atributos que sero mais importantes para a construo do nosso modelo de predio. 

**Primeira Soluo : Mtodo de Filtro**

```python
# geranda a corrao entre os atributos usando a correlao de Pearson
# DataFrame Train
plt.figure (figsize = (12,10)) 
cor = df_train.corr () 
sns.heatmap(cor, annot = True, cmap = plt.cm.Reds) 
plt.show ()
```

```python
# Correlao com a varivel alvo 
cor_target = abs(cor["survived"])
# Selecionando recursos altamente correlacionados, acima de 0.3
relevant_features = cor_target[cor_target> 0.3] 
relevant_features
```

**observao** : O primeiro mtodo sugere a manuteno dos atributos *pclass* e *sex* 

**Segunda Soluo** : Mtodo *Embedded*

```python
X = df_train.drop("survived", axis=1)   
y = df_train["survived"]

reg = LassoCV()
reg.fit(X, y)
print("Melhor alpha usando LassoCV: %f" % reg.alpha_)
print("Melhor score usando LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```
